## Run

```python
CONFIG = "configs/deepseek_openai.yaml"
MODE = None  # set to "full" or "retrieval" to override the config
INPUT_CSV = None
INPUT_JSONL = None
OUTPUT_JSONL = None
REPORT_FILE = None

cmd = ["python", "scripts/run_pipeline.py", "--config", CONFIG]
if MODE:
    cmd += ["--mode", MODE]
if INPUT_CSV:
    cmd += ["--input-csv", INPUT_CSV]
if INPUT_JSONL:
    cmd += ["--input-jsonl", INPUT_JSONL]
if OUTPUT_JSONL:
    cmd += ["--output-jsonl", OUTPUT_JSONL]
if REPORT_FILE:
    cmd += ["--report-file", REPORT_FILE]

import subprocess
print(" ".join(cmd))
subprocess.run(cmd, check=True)
```
